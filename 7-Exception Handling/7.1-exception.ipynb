### Understanding Exception

### What are Exception?

##### Exception are events that distrub the normal flow of a program. they occur when an error is encounterd during program execution. Common Execution include:

##### ZeroDivisonError : Dividing By Zero
##### FileNotFoundError : File Not Found
##### ValueError : Invalid Value
##### TypeError : Invalid Type

```python
try:
    num=int(input("enter the number"))
    result=10/num
except ValueError:
    print("This is not a valid Number")
except ZeroDivisionError:
    print("Enter Denominator Greater than Zero")
except Exception as ex:
    print(ex)
```

```python
# Try,except,else block 
try:
    num = int(input("enter a number:"))
    result = 10/num
except ValueError:
    print("This is not a valid Number!")
except ZeroDivisionError:
    print("Enter Denominator Greater than Zero!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
```

```python
# try,except,else,finally
try:
    num = int(input("enter a number:"))
    result = 10/num
except ValueError:
    print("This is not a valid Number!")
except ZeroDivisionError:
    print("Enter Cant divide by Zero!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
finally:
    print("execution complete.")
```
