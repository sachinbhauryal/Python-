```python
# if statement

age = 20

if age>= 18:
    print("you can vote.")
```

```python
# else
# the else statement execute the code if the condition in the if statement is false
age =14 
if age>=18:
    print("you can vote.")
else:
    print("you cannot vote.")
```

```python
# elif
# the elif condition allow you to check multiple conditions. It stands for "else-if"
age = 80 
if age<13:
    print("you are child")
elif age<40:
    print("you have adult now.")
else:
    print("you are old now.")
```

```python
# Nested condition statement

num = int(input("enter the number:"))

if num>0:
    print("the number is poistive.")
    if num%2 == 0 :
        print("the number is even")
    else:
        print("the number is odd")
else:
    print("the number is the negative number.")
```

```python
# Assignment
# Simple calculator

num1 = float(input("Enter the first number:"))
num2 = float(input("enter the second number:"))
operator = input("enter operator ('+', '-' ,'*' ,'/' ):")

if operator == '+':
    result = num1 + num2 
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    result = num1 / num2 
else:
    result = "Invalid Operation."

print(f"Result : {result}")
```
