### Indentation 

Indentation in python is used to define the structure and hierarcy of the code. unlike many other programming language that uses braces{} to delimit block of code, Pyhon uses identation to determine the grouping of the statement. this means that all the statement within the block must be indented at the same level.




```python
# identation
# python uses indentation to define a block of code. The standard indentation is 4 spaces or 1 tab.

age = 25
if age >= 18:
    print("You are an adult.")
else:
    print("you are minor.")
```
