# Data Manipulation and Analysis with Pandas

Data manipulation and analysis are key task in any data science or data analysis project.Pandas provide awide range of function for data manipulation and analysis,making it easier to clean, transorm and extract insight from data.

```python
import pandas as pd
```

```python
df = pd.read_csv('nwe.csv')
df.head(6)
```

```python
df.tail(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
# Handling Missing value
df.isnull()
```

```python
df.isnull().any(axis=1)
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
dfFilled = df.fillna(0)
dfFilled
```

```python
# filling missing value with the mean of the column
df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes
```

```python
#Renaming Column 
df = df.rename(columns={'Date':'Sales Date'})
df.head(5)
```

```python
# change Datatypes
df['Value_new'] = df['Value'].fillna(df['Value'].mean())
df.tail(20)
```

```python
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.tail(20)
```

```python
df['New_sales'] = df['Sales'].apply(lambda x:x**2)
df.head(4)
```

```python
## Data aggregation and grouping

grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
Grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(Grouped_sum)
```

```python
# Aggregate Multiple Functions
Grouped_aggregate = df.groupby('Region')['Value'].agg(['mean','sum','count'])
Grouped_aggregate
```

```python
# Merging and joining Dataframes
# create sample Dataframes

df1 = pd.DataFrame({'key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'key':['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#Merge DataFrame on the 'Key Column'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
