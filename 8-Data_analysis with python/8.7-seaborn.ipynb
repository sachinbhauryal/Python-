## Data Visualizatin With Seaborn :-

Seaborn is the python visualization library based on Matplotlib that provide the high level interface for drawing attractive and informative statistical graphics.Seaborn help in creating complex complex visualization with just few lines of codes.In this lesson we will cover the basic of seaborn,including creating various types of plots and customizing it.

```python
import seaborn as sns
```

```python
# Basic plooting with seaborn 
tips = sns.load_dataset('tips')
tips
```

```python
## create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill VS tip")
plt.show()
```

```python
# Line Plot 
sns.lineplot(x='size',y='total_bill',data=tips,marker='o')
plt.title("line plot of total bill by size")
```

```python
# Categorial Plots

#Bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of total bill by day")
plt.show()
```

```python
#Box plot

sns.boxplot(x='day',y='total_bill',data = tips)
```

```python
# Violin plot

sns.violinplot(x='day',y='total_bill',data = tips)
```

```python
## Histogram 
sns.histplot(tips['total_bill'],bins = 10,kde =True)
```

```python
# KDE plot
sns.kdeplot(tips['total_bill'],fill = True)
```

```python
# pair plot
sns.pairplot(tips)
```

```python
# Heat map
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# plot ttoal sales by product

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total sales by product')
plt.xlabel('Product')
plt.ylabel('Total sales')
plt.show()
```

```python
# total sale by region
plt.figure(figsize=(8,5))

sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total sales by Region')
```
