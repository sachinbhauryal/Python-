#### Pandas Dataframe and series

##### pandas is the powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structure: Series and Datframes.A Series is the one dimensional array like object,while a dataframe is a two-dimenstional,size-mutual,and potentially hetrogeneous tabular data structure with labeled axes(rows and colums).

```python
## Series 
# A pandas series is a one dimensional array-like object that can hold any data type. 

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# create a series from to dictonary
data = {'a': 1, 'b': 2 ,'c': 3,'d':4}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30,40]
index = ['a','b','c','d']
pd.Series(data,index = index)
```

```python
# DataFrame
# Create a datagram from a Dictonary of list 

data = {
    'Name' : ['krish','john','jack'],
    'Age' : [23,12,64],
    'City': ['Banglore','New York','Florida']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# Create a list of data frames from a list of Dictonaries
data = [
     {'Name':'Krish','Age':32,'City':'Bangalore'},
     {'Name':'Sachin','Age':21,'City':'Mumbai'},
     {'Name':'Ramesh','Age':30,'City':'jaipur'}
]
dff = pd.DataFrame(data)
print(dff)
print(type(dff))
```
