### The Map() Function in Python

The map() function applies a given function to all items in an input list( or any other iterable) and return a map object(an iterator).this is practically useful for transforming data in a list comprehensively.


The map() function applies a function to each item in an iterable (like a list, tuple, etc.) and returns a map object.

```python
def square(x):
    return x**2

square(10)
```

```python
number = [1,2,3,4,5,6,7]
list(map(square,number))
```

```python
# Lambda Function with map

number = [1,2,3,4,5,6]
list(map(lambda x:x**3,number))
```

```python
# Map multiple iteration

num1 = [1,2,3]
num2 = [4,5,6]

added = list(map(lambda x,y : x+y,num1,num2))
print(added)
```

```python
# Map to convert list of string into integers

str_number = ['1','2','3','4','5']
int_number = list(map(int,str_number))
print(int_number)
```

```python
words = ["apple","mango","banana"]
upper_words = list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(person):
    return person["name"]

people =[
    {'name':'krish','age':32},
    {'name':'sachin','age':33}
]

list(map(get_name,people))
```
