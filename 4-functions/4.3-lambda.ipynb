### Lambda Fuctions in Python

##### Lambda are small anonymus fuction defined using the Lambda Keyword. They can have any number of arguments but only one expression.They are commonly used for short operations or as arguments to higher-order Functions.

```python
# Syntax

lambda arguments: expression 
```

```python
def addition(a,b):
    return a+b
```

```python
addition(3,2)
```

```python
add = lambda a,b:a+b
print(add(4,1))
print(type(add))
```

```python
def even(num):
    if num%2==0:
        return True

even(24)
```

```python
even1 = lambda num:num%2==0
print(even1(23))
```

```python
def addit(a,b,c):
    return a+b+c

addit(1,2,3)
```

```python
addit1 = lambda a,b,c:a+b+c

addit1(1,2,4)
```

```python
## map - Applies a function to all items in the list.
number = [1,2,3,4,5,6]
def square(number):
    return number**2

square(4)
```

```python
list(map(lambda x:x**2, number))
```
