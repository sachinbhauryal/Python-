```python
# why function?

num = 24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_or_odd(num):
    """ this function find even or odd"""
    if num%2 == 0:
        print("The number is even")
    else:
        print("The number is odd")

```

```python
# Call the Function

even_or_odd(23)
```

```python
# Function with multiple parameters

def add(a,b):
    return a+b

result = add(34,7)
print(result)
```

```python
# Default Parameter 

def greet(name):
    print(f" Hello {name},Welcome to the paradise") 

greet("sachin")
```

```python
def greet(name = "Guest"):
    print(f" Hello {name},Welcome to the paradise") 

greet()
```

```python
# Variable length arguments
# Positional and key words argument


## Positional Argumnets

def printNumbers(*args):
    for number in args:
        print(number)
    
printNumbers(1,2,34,1,3,42,12,"sachin")
```

```python
# Keywords arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
        
```

```python
 print_details(Name = "Sachin",Age = 23, Country = "India")
```

```python
def print_details( *args,**kwargs):
    for  val in args:
        print(f"Positional Arguments : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"krish",Name="Sachin",age=23,Country="India")
```

```python
# Retuen statement

def multiply(a,b):
    return a*b

product = multiply(3,10)
print(product)
```

```python
# Return multiple Parameters
def multiply(a,b):
    return a*b,a,b

product = multiply(3,10)
print(product)
```
