###  File Operation Read and write File

File Handling is a crucial part of any programming language.Python provides build in function and method to read from and write to files,both text and binary.The lesson will cover the basic of file handling including reading and writing text  file and binary files.

```python
##  Read a whole files

with open('example.txt',mode='r') as file:
    content = file.read()
    print(content) 
```

```python
# read a file line by line

with open ('example.txt',mode='r') as file:
    for line in file:
        print(line)
```

```python
with open ('example.txt',mode='r') as file:
    for line in file:
        print(line.strip())  #strip is used to remove the new line space
```

```python
# Writing a file (overwriting)

with open ('example.txt','w')as file:
    file.write('Hello world!\n')
    file.write('this is a new line.')
```

```python
# write a file (without overwriting)

with open('example.txt', mode='a') as file:
    file.write('Append operation is taking place')
```

```python
# Writing alist of line in to a file

lines = ['First line.\n','Second line.\n','Third line.\n']
with open ('example.txt',mode='a') as file:
    file.writelines(lines)
```

```python
### Binary files

# Writing to a binary files

data = b'\x06\x01\x03\x04\x05'

with open('data.bin', 'wb') as file:
    file.write(data)

print("Binary data written successfully!")
```

```python
# reading a binary file 

with open('data.bin',mode='rb') as file:
    content = file.read()
    print(content)
```

```python
# Read the content from the source text file and write to a destination text file

with open('example.txt',mode='r')as source_file:
    content = source_file.read()

with open('destination.txt',mode='w')as destination_file:
    destination_file.write(content)
```

```python
# Read a text file and  count the no of line,words, and characters.

# Counting lines,words,and character in a text file

def count_text_file(file_path):
    with open(file_path, 'r')as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count,word_count,char_count

file_path = "example.txt"
lines, words ,character = count_text_file(file_path)
print(f'lines:{lines}, Words :{words},Character:{character}') 
```

#### The w+ mode in python is used to open a file for both reading  and writing.if the file does not exist it will be created.if the file exists,its content is truncated(i.e,the file is overwritten.)

```python
# Writing and then reading the file 

with open('example.txt','w+')as file:
    file.write("Hello users!, how are you \n")
    file.write("see you soon!")

    # Move the cursor to the beginnig

    file.seek(0)

    #Read the content of the file
    content = file.read()
    print(content)
```

```python
path = 'example.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists.")
else:
    print(f"The path '{path}' does not exists.")
```

```python
# Checking if the path is a file or directory

import os

path = 'example.txt'

if os.path.isfile(path):
    print(f" The path '{path}' is a file")
elif os.path.isdir(path):
    print(f"the path '{path}'is a directory")
else:
    print(f"The file '{path}' is neither a file nor a directory")
```

```python
# getting the absolute path 
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
