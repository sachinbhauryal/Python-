###  File OPeration Read and write File

File Handling is a crucial part of any programming language.Python provides build in function and method to read from and write to files,both text and binary.The lesson will cover the basic of file handling including reading and writing text  file and binary files.

```python
##  Read a whole files

with open('example.txt',mode='r') as file:
    content = file.read()
    print(content) 
```

```python
# read a file line by line

with open ('example.txt',mode='r') as file:
    for line in file:
        print(line)
```

```python
with open ('example.txt',mode='r') as file:
    for line in file:
        print(line.strip())  #strip is used to remove the new line space
```

```python
# Writing a file (overwriting)

with open ('example.txt','w')as file:
    file.write('Hello world!\n')
    file.write('this is a new line.')
```

```python
# write a file (without overwriting)

with open('example.txt', mode='a') as file:
    file.write('Append operation is taking place')
```

```python
# Writing alist of line in to a file

lines = ['First line.\n','Second line.\n','Third line.\n']
with open ('example.txt',mode='a') as file:
    file.writelines(lines)
```

```python
### Binary files

# Writing to a binary files

data = b'\x06\x02\x03\x04\x05'

with open('data.bin', 'wb') as file:
    file.write(data)

print("Binary data written successfully!")
```

```python
# reading a binary file 

with open('example.bin',mode='rb') as file:
    content = file.read()
    print(content)
```
