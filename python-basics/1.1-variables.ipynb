```python
## declaration and assigning variables: 
age = 25
name = "Sachin Bhauryal"
height = 6.5
is_student = True

# printing the variables:

print("Age :", age)
print("Name :", name)
print("Height :", height)

```

```python
# understanding variable types:
# python is the dynamically typed language, so we don't need to declare the type of variable, it will be inferred by the value assigned to it.
age = 25   #integer
name = "Sachin Bhauryal"  #string
height = 6.5  #float
is_student = True #boolean
```

```python
# Type checking and Conversion

print(type(age))  #<class 'int'>

# Typer Conversion:
age_str = str(age)  # converting integer to string
print(type(age_str))  #<class 'str'>
```

```python
# Dynamic Typing

# python allow the type of a variable to change as  the program executes.
var = 10
print(var,type(var))

var = "Now I'm a string"
print(var,type(var))

var = 3.14
print(var, type(var))
```

```python
# Simple calculator

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
Product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", Product)
print("Quotient :", quotient)
```
