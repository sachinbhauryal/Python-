### Nominal/OHE Encoding
one hot encoding also known as nominal encoding,is a technique used to represent categorial data as numerical data which is more suitable for machine learning algorithm.in this each category is represented as a binary vector where each bit correspond to a unique category.For Example:- If we have a categorical values "color" with three possible value(red,green,blue)we can represent it as one hot encoding as follow:

1. Red[1,0,0]
2. Green[0,1,0]
3. Blue[0,0,1]

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a sample dataframe:

df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Green', 'Red', 'Blue']})
```

```python
df.head(5)
```

```python
# create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['Color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
new_df = pd.concat([df,encoder_df],axis=1)
```

### 2. One-Hot Encoding

Use this when the categories do not have any order.

```python
from sklearn.preprocessing import LabelEncoder
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(tips[['day']])
print(encoded.shape)
```

```python
print(encoder.get_feature_names_out(['day']))
```

```python
import pandas as pd

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(['day'])
)

print(encoded_df.head())
```
