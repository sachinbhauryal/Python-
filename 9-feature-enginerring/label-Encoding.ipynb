### Label Encoding

Label encoding and Ordinal Encoding are two techinque used to encode categorical data as numerical data.

Label encoding involves assigning a unique numerical label to each category in the variable. the label are usually assign in the alphabetical order or based on the frequency of the categories.For Example: If we have a categorical variable "color" with threee possible values(red,green,blue)we can represent it using label encoding  as follows:

1. Red:1
2. Green:2
3. Blue:3

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Green', 'Red', 'Blue']})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
```

```python
lbl.fit_transform(df[['Color']])
```

```python
lbl.transform([['Red']])
```

```python
lbl.transform([['Green']])
```

```python
lbl.transform([["Blue"]])
```

### Ordinal Encoding 

It is used to encode categorical data that have an intrinsic order or ranking. in this technique each category is assigned a numerical value based on its position in the order. For Example:- if we have a categorical variable "eduction level" with four possible value (high school,collage,graduate,post-graduate),we can represent it using ordinal encoding as follows:

1. High School:1
2. College:2
3. Graduate:3
4. post-Graduate:4

```python
# ordinal Encoding

from sklearn.preprocessing import OrdinalEncoder
```

```python
# Create a sample dataframe with an ordinal variable
df = pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
# create an istance of Ordinal Encoder and perform fit_transform
```

```python
encoder = OrdinalEncoder(categories =[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

### Target Guided Ordinal Encoding :-

It is the technique used to encode categorical based on the relationship with the target variable.this encoding technique is useful when we have a categorical variable with a large number of unique categories, and we want to use this variable as a feature in our machine.

In target Guided Ordinal Encoding, we replace each category in the categorical variable with a numerical value based on the mean and median of the target variable for the category.This create a monotonic relationship between the categorical variable and the target variable,which can improve the pridictive power of our model.


```python
import pandas as pd
#create a sample dataframe with a categorical variable and a target variable
df=pd.DataFrame({
    'city':['New York','Londan','Paris','Tokyo','New York', 'Paris'],
    'Price':[200,150,300,250,180,320]
}) 
```

```python
df
```

```python
mean_price=df.groupby('city')['Price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['city','city_encoded']]
```
