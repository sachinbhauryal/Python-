## Sets

##### Sets are build in data type in python used to store collection of unique items. the are unordered, meaning that the element do follow the specific order, and they do not aloow duplicate elements.sets are useful for membership tests, eliminating duplicate entries and performing mathmetical set operation like union,intersection,difference and symetic difference.

```python
# create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(my_empty_set)
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,4,7,8])
print(my_set)
print(type(my_set))
```

```python
unique = {1,2,4,1,7,2,1}
print(unique)
```

```python
# Basic set operation
# Adding and Removing Elements

new_set = {1,2,3,4,5}
print(new_set)
new_set.add(6)
print(new_set)
new_set.add(7)
print(new_set)
new_set.add(7)
print(new_set)
```

```python
# Removing the element from the set

new_set.remove(6)
print(new_set)
```

```python
new_set.discard(11) #remove the element if it is a member.
print(new_set)
```

```python
# pop element
removed_element = new_set.pop()
print(removed_element)
print(new_set)
```

```python
# clear all the element
new_set.clear()
print(new_set)
```

```python
# set membership test

new_set = {1,2,3,4,5}
print(2 in new_set)
print(10 in new_set)
```

```python
# Mathematical operation
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# Union
Union_set = set1.union(set2)
print(Union_set)

# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2) # it will update the set1 by the element which are common in both the set
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# Difference
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
#Symmetric difference
set1.symmetric_difference(set2)  # unique element from both set1 and set2
```

```python
# set methods


# is subset
set1 = {1,2,3,4,5}
set2 = {3,4,5}

print(set1.issubset(set2))

# is super set
print(set1.issuperset(set2))
```

```python
lst = [1,2,2,3,4,4,2,3,5]
print(set(lst))
```

```python
# counting unique words in the text

text = "in this tutorial we are discussing about sets in we are "
words = text.split()
print(words)

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
