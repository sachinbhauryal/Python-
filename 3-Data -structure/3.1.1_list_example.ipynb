### Example 1 : Manage a TO DO List

  #### Create a TO DO Lists to manage track of the task.

```python
To_do_list = ["wake up at 4AM","Do Exercise","GO for Running","Complete the Lecture"]

#Adding a tasks.
To_do_list.append("take less break")
To_do_list.append("don't take evening sleep")

#Removing a complete tasks
To_do_list.remove("wake up at 4AM")
To_do_list.remove("Complete the Lecture")

# Reminder 
if "GO for Running" in To_do_list :
    print("HURRY UP! , Go for the running")

print(f" Task Remaing To Complete :-")
for task in To_do_list:
    print(f"-{task}")
```

### Example :- Organizing Student Grades 

#### Create a list to store and calculate average grade for student.

```python
# Organizing student grade

grade = [85,92,79,97]

# Adding a new grade
grade.append(91)
print(grade)

# Calculating the average grade
average_grade = sum(grade)/len(grade)
print(average_grade)

# Finding the highest and the lowest grade
higest_grade = max(grade)
Lowest_grade = min(grade)

print(f"Higest Grade : {higest_grade}")
print(f"Lowest Grade : {Lowest_grade}")
```

### Example :- Managing the Inventory

#### Use a list to manage inventry item in the store

```python
# Managing the inventry 
inventory = ["apple","mango","gauva","grapes"]

# Adding an item in the stock

inventory.append("kiwi")
inventory.append("orange")

print(inventory)

# Removing the item which is out of stock 

inventory.remove("apple")
print(inventory)

# Checking if item is in the stock

item = "gauva"
if item in inventory:
    print(f"{item} is in the stock")
else:
    print(f"{item}is out of stock")
```

### Example 4 :- Collecting user Feedbacks

##### use a list to collect user feedbacks

```python
# collecting user feedbacks 

feedback = ["Great service!","very satisfied","Could be better","Excellent experience"]

# Adding new feedbacks 
feedback.append("not happy with the service")

# counting specific feedbacks
positive_feedback_count = sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())
print(f" positive feedbacks count : {positive_feedback_count}")

# print all feedbacks
for review in feedback:
    print(f"-{review}")
```
