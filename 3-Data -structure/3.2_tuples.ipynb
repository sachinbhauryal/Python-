## Introduction to Tuples 
#### Tuples  are orderes collection of items that are immutable.They are similar to lists,but there immutablity made them different


```python
empty_tuples = ()
print(empty_tuples)
print(type(empty_tuples))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
print(type(numbers))
numbers
```

```python
num = list[(1,2,3,4)]
print(type(num))
num
```

```python
mixed = (1,"sachin",6.1,True)
mixed
```

```python
# Accessing Tuples element

numbers
```

```python
print(numbers[1])
print(numbers[-1])
```

```python
# Slicing
print(numbers[0:4])
print(numbers[:5])
print(numbers[5:])
print(numbers[::-1])
```

```python
# Tuples Operations

Concatenation = numbers +  mixed
print(Concatenation)
```

```python
mixed*3
```

```python
numbers*3
```

```python
# Immutable nature of Tuples
# Tuples are Immutable means there element cannot be changed once they are assigned

lst = [1,2,3,4,5]
print(lst)
lst[0]= "krish"
print(lst)
```

```python
num = (1,2,3,4)
print(num)
num[1] = "sachin"
print(num)
```

```python
# Tuples Methods
numbers
```

```python
print(numbers.count(1)) # how many time the element is present in the tuple
print(numbers.index(3)) # print the index of the given number
```

```python
# Packing and Unpacking Tuples 
## Packing
Packed_tuples = 1,"hello",3.12
print(Packed_tuples)
```

```python
## Unpacking a tuple
a,b,c = Packed_tuples
print(a)
print(b)
print(c)
```

```python
# Unpacking with *
numbers1 = (1,2,7,4,8,9)
first,*middle,last = numbers1
print(first)
print(middle)
print(last)
```

```python
# Nested list
lst = [[1,2,3,4],[5,6,7,8],[1,"hello",3,14,"c"]]
print(lst[1][2])
print(lst[2][0:3])
```

```python
# a nested list can also contain a tuple in it 
lst = [[1,2,3,4],[5,6,7,8],(1,"hello",3,14,"c")]
lst[2][0:3]
```

```python
# Nested Tuples 
nested_tuples = ((1,2,3),(4,5,6),("a","b","c"),(True,False))
print(nested_tuples[3])
print(nested_tuples[2][2])
```

```python
# iterating over nested_tuples
for sub_tuple in nested_tuples:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
