### Dictionaries 

##### Dictionaries are unordered collection of items.they store data in key value pairs. Keys must be unique and immutable (e.g., string, numbers and tuples),while value can be of any type.

```python
# Creating Dictionaries
empty_dictionaries = {}
print(type(empty_dictionaries))
```

```python
empty_dictionarie = dict()
print(type(empty_dictionarie))
```

```python
student = {"Name":"Sachin","age":32,"grade":24}
print(student)
```

```python
student = {"Name":"Sachin","age":32,"Name":24} # in this case the name will not print 2 time it will be updated.
print(student)
```

```python

student = {"Name":"Sachin","age":32,"grade":"A"}
print(student)
```

```python
# Accessing Dictionary element 

print(student["Name"])
print(student["age"])

# Accessing using get() method
print(student.get("grade"))
print(student.get("lastName"))
print(student.get("lastName","Not available"))
```

```python
# Modify Dictionaries element
# Dictionaries are mutable,so you van add,update, or delete elements.

print(student)
```

```python
student["age"] = 33  ## updated  value of the key
print(student)
student["Address"] ="india" ## Add a new Key and Value
print(student)
```

```python
del student['grade']  #delete key and value pair

print(student)
```

```python
# Dictionaries method

keys = student.keys()
print(keys)
values = student.values()
print(values)
both = student.items()
print(both)
```

```python
# Shallow copy

student_copy = student
print(student)
print(student_copy)
```

```python
student["Name"] = "Sachin"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy()
print(student_copy1)
print(student)
```

```python
student["Name"]="krish" # this is the shallow copy in which no change in the previous copy will affect it.
print(student)
print(student_copy1)
```

```python
# Iteration over dictionaries
# you an use loops to iteration

# Iteration over keys

for keys in student.keys():
    print(keys)
```

```python
# Iteration over values

for value in student.values():
    print(value)
```

```python
# Iteration over keys values pairs
for key,value in student.items():
    print(f"{key} : {value}")
```

```python
# Nested Dictionaries

students ={
    "student1" : {"name":"rahul","age":22},
    "student2" : {"name":"amit","age":21},
    "student3" : {"name":"sohan","age":30}
}
print(students)
```

```python
# acessing nested dicitionaries element

print(students["student2"]["name"])
print(students["student3"]["name"])
```

```python
# iterating over nested dictionaries
for key,value in students.items():
    print(f"{key}:{value}")
    for inside_key,inside_value in value.items():
        print(f"{inside_key} : {inside_value}")
```

```python
# Dictionaries Comprehension

square = {x:x**2 for x in range(6)}
print(square)
```

```python
# Conditional dictionaries comprehension

even_square = {x:x**2 for x in range(1,11) if x%2==0}
print(even_square)
```

```python
# Practice Example

# Use a dictionary to count the frequency of element in the dictionary

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency= {}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]= 1

print(frequency)
```

```python
# merging 2 dictionaries

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
merged_dictonaries = {**dict1,**dict2}
print(merged_dictonaries)
```
